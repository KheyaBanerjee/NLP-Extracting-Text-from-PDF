```python
# Import libraries
import PyPDF2

from nltk.tokenize import word_tokenize
```

### Read sample file

```python
# Open and read pdf
sample_pdf = open(r"C:\Users\KhBanerjee\OneDrive - American Bureau of Shipping\Desktop\AcronymsList.pdf", mode="rb")
pdfdoc = PyPDF2.PdfReader(sample_pdf)
```

### Displaying the file contents

```python
pdfdoc.metadata
```

```python
len(pdfdoc.pages)
```

```python
#res = []
res = ''
for i in range(len(pdfdoc.pages)):
    current_page = pdfdoc.pages[i]
    # DO NOT CONVERT TO LIST ..... it will include \n
    #res.append(current_page.extract_text())
    res += current_page.extract_text()
print(res)
```

```python
print(res[0])
```

```python
# Use tokenization instead to make a list
print(type(word_tokenize(res)))
print(word_tokenize(res))
```
